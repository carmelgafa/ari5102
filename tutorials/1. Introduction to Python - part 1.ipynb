# What is Python?

Python is an interpreted, object-oriented, high-level programming language. Its high-level built in data structures makes it very attractive for Rapid Application Development, as well as for use as a scripting or glue language to connect existing components together. Python's simple, easy to learn syntax emphasizes readability and therefore reduces the cost of program maintenance.

##Advantages
*   <u>Easy to Read, Learn and Write</u>: Python is a high-level programming language that has English-like syntax. This makes it easier to read and understand the code.
*   <u>Improved Productivity</u>: Due to the simplicity of Python, developers can focus on solving the problem. They donâ€™t need to spend too much time in understanding the syntax or behavior of the programming language.
*   <u>Interpreted Language</u>: Python is an interpreted language which means that Python directly executes the code line by line without the need for compiling it first.
*   <u>Free and Open-Source</u>: Python is free to use and distribute.
*   <u>Vast Libraries Support</u>: The standard library of Python is huge, but there is also the *Python package index (PyPi)*, which consists of over 200,000 packages free to download and use. So, you can find almost all the functions needed for your task.

##Disadvantages
*   <u>Slow</u>: Since Python is an interpreted programming language, the execution of python code can be slow compared to compiled statically typed programming languages, such as C and Fortran.
*   <u>Decentralized</u>: Different environments, packages and documentation spread out at different places.

### Type casting
Type casting the process of converting the value of a single data type (such as an integer, float, or double into another data type.

When a floating point number is converted to an integer, its decimal part is discarded.

When an integer is converted to a boolean variable, the boolean variable takes the value `True` is the integer is different than zero.

```python
y = 2.8
print(y, type(y))

y = int(y)
print(y, type(y))
```

```python
x = '5.0'
z = y + float(x)
print(z, type(z))
```

```python
x = -5
print(bool(x), type(bool(x)))

y = 0
print(bool(y), type(bool(y)))
```

## Comparisons

Comparisons are operators which evaluate properties of their operands and always return a boolean value (`True` or `False`)

Common comparison operators:
* `<`
* `>`
* `==`
* `>=`
* `<=`
* `!=`

```python
x = 1 >= 2
print(x, type(x))
```

```python
x = 2 == 2.0
print(x, type(x))
```

```python
x = 2 < 5 > 3
print(x, type(x))

x = 2 < (5 > 3) # why is this False?
print(x, type(x))
```

`is` and `is not` comparisons check whether two operands point to the same object. They **do not** check just equality.

```python
x = 1
y = 1
print(x is y) # both are integers, and thus they point to the same object

z = 1.0
print(x is z) # one is integer the other one is float, and thus they do not show to the same object
print(x == z)
```

### Boolean operations

* `and`

  * True `and` False = False
  * False `and` True = False
  * False `and` False = False
* `or`
  * True `or` True = True
  * True `or` False = True
  * False `or` True = True
  * False `or` False = False

  * `not` True = False
  * `not` False = True

```python
print(2 < 3 and 5 > 10)
```

```python
print(not 5 < 8)
```

```python
print(2 < 3 or 5 > 10)
```

## Strings

```python
s = "Konstantinos Makantasis"

length_of_s = len(s)
print('length of s is:', length_of_s)

fifth_letter = s[4] # be careful: the index for the first letter is 0
last_letter = s[-1] # index -1 returns the last letter of the string
third_to_seventh_letters = s[2:7] # the letter with index '7' is not included
print('the fifth letter is:', fifth_letter)
print('the last letter is:', last_letter)
print('the third to seventh letters are:', third_to_seventh_letters)
```

### String replacement

```python
print(s.replace('Konstantinos', 'Kostas'))
print(s) # the replacement above does not alter the value of s

s = s.replace('Konstantinos', 'Kostas') # to alter the value of s we need to make a new assignment
print(s)
```

### String concatenation

```python
firstname = 'Konstantinos'
surname = 'Makantasis'
fullname = firstname + ' ' + surname
print('firstname:', firstname)
print('surname:', surname)
print('fullname:', fullname)
```

### String formating
* It enables combining strings
* It is bery useful for printing

```python
s = '{} {}'.format('Konstantinos', 'Makantasis')
print(s)
```

```python
s = '{1} {0}'.format('Konstantinos', 'Makantasis')
print(s)
```

```python
pi = 3.14159
print('{} equals {:.2f}'.format('Pi', pi)) # it prints only the first two decimals
```

## Lists
* A list is similar to a string in that it's a collection of objects
* In Python a list can contain objects of any type

```python
empty_list = []
a = [1, 2, 3]
print(empty_list, type(empty_list))
print(a, type(a))
```

```python
a = ['one', 'two', 'three']
print(a, type(a))
```

```python
a = [1, 'two', 3.0]
print(a, type(a))
print('The first list element is {} with type {}: '.format(a[0], type(a[0]))) # we access the elements of a list using the following syntax: list_name[index]
print('The second list element is {} with type {}: '.format(a[1], type(a[1])))
print('The third list element is {} with type {}: '.format(a[2], type(a[2])))
```

### Nestted lists
* A nested list is a list of lists

```python
a = [[1, 2, 3], [1, 'two', 3.0]]
print(a)
```

```python
a1 = a[0] # this is the first list of a
a2 = a[1] # this is the second list of a
print(a1)
print(a2)
```

To access a specific element of a specific list we use two indices
* The first index indicates the list of interest
* the second the element of interest

```python
a1_last = a[0][-1] # the last element of the first list. We use the index -1 to access the last element of a list.
a1_second_last = a[0][-2] # the second last element of the first list.
a2_3 = a[1][2] # the third element of the second list
print(a1_last)
print(a1_second_last)
print(a2_3)
```

For printing all the lements of a list we use `:`

```python
all_elements_of_second_list = a[1][:] # all the elements of the second linst in list a
print(all_elements_of_second_list)
```

### Range
* Range is used to automatically create list of integers.

```python
start = 2
stop = 20
step = 3
a = list(range(start, stop, step)) # the value stored in stop is not included in the list
print(a)
```

### List modification
There are four main operators for list modification
* `append` : adds an elements at the end of the list
* `insert` : adds an element at a specific position
* `remove` : removes the first element withe the given value
* `pop` : removes the element at a given position

```python
a.append(20)
print(a)
```

```python
a.insert(1, 3) # inserts the value of 3 as the second elements
print(a)
```

```python
a.remove(3)
print(a)
```

```python
a.pop(5) # removes the element at the 5th position
print(a)
```
